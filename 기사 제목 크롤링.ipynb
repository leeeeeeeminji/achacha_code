```python
# -*- coding: utf-8 -*-
"""기사 제목 크롤링

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1tzA5PU_uraphg8g-qqfg0sFxAOd-Lntm

<기사 제목 크롤링>

한남대학교 대표팀 '앗차차'의 ppt 11번째 슬라이드에 들어가는 워드클라우드를 위한 기사제목 크롤링 설명입니다.
"""

import requests  #웹페이지에서 HTTP 요청을 보내 원하는 HTML 정보를 가져오는 모듈
from bs4 import BeautifulSoup #BeautifulSoup 불러오기

search_keyword = '어린이 차량 갇힘' #크롤링 검색어 설정

#페이지 자동 넘어가게 url끌어오기
url_format = 'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EC%96%B4%EB%A6%B0%EC%9D%B4%20%EC%B0%A8%EB%9F%89%20%EA%B0%87%ED%9E%98&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=11&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start={0}'


article_count = 0
titles = []

while article_count < 1500:
  r = requests.get(url_format.format(article_count))#request로 url가져오기
  soup = BeautifulSoup(r.text, 'html.parser') #문자열 BeautifulSoup 으로 읽어주기
  news_titles = soup.select('.sp_nnews .news_wrap .news_tit') #기사 제목 호출

  for title in news_titles:
    titles.append(title['title']) #기사 제목 추가
    article_count += 1  #기사 제목 개수 세기 위해 반복문 돌 때 마다 article_count에 +1씩 더해주기 
    
for title in titles:  #기사제목 출력
  print(title)
```
